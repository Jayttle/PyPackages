```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix
plt.rcParams['font.sans-serif'] = ['Kaiti']
plt.rcParams['axes.unicode_minus'] = False
import torch
```

# 读取数据

```python
# 获取batch
def get_batch(df, target):
    df_copy = df.copy()
    df_copy['change_point'] = (df_copy[target] != df_copy[target].shift(1)).fillna(False).astype(int)
    df_copy['batch'] = df_copy['change_point'].cumsum().astype('int')
    return df_copy['batch']
```

```python
# 读取原始数据
df = pd.read_csv('./data/B18_DGTPM_20240829_.csv')
dfs = pd.read_csv('./data/B18_DGTPM_RD4757_20240903_20240910.csv')
# 提取4757数据
moto_4757_abnorm = df[df['Sensor_ID'] == '4757-1.Sensor']
moto_4757_norm = dfs[dfs['motor_id'] == '4757-1']
# 统一columns
moto_4757_abnorm.columns = moto_4757_abnorm.columns.str.lower()
# 转换日期类型
moto_4757_norm['datetime'] = pd.to_datetime(moto_4757_norm['datetime'])
moto_4757_abnorm['datetime'] = pd.to_datetime(moto_4757_abnorm['datetime'])
# 提取固定列
moto_4757_abnorm = moto_4757_abnorm[['datetime', 'motor_id', 'motor_run', 'motor_speed', 'motor_temperature', 'motor_torque', 'sensor_state']]
moto_4757_norm = moto_4757_norm[['datetime', 'motor_id', 'motor_run', 'motor_speed', 'motor_temperature', 'motor_torque', 'sensor_state']]
# 增加异常标签
moto_4757_abnorm['label'] = 1
moto_4757_norm['label'] = 0
# 合并数据
moto_4757_whole = pd.concat([moto_4757_norm[['datetime', 'motor_id', 'motor_run', 'motor_speed', 
                                             'motor_temperature', 'motor_torque', 'sensor_state', 'label']], moto_4757_abnorm], axis=0)
# 获取batch
moto_4757_whole['batch'] = get_batch(moto_4757_whole, 'motor_run')
```

```python
# 获取batch标签
moto_4757_norm['batch'] = get_batch(moto_4757_norm, 'motor_run')
moto_4757_norm_stop = moto_4757_norm.groupby(['batch', 'motor_run'])['datetime'].count().reset_index()
moto_4757_norm_stop = moto_4757_norm_stop[moto_4757_norm_stop['motor_run'] == 0]
# 获取停机批次
moto_4757_norm_stop = moto_4757_norm_stop[moto_4757_norm_stop['datetime'] > 750]
# 获取停机结束时间
moto_data_4757_norm_stop_time = (moto_4757_norm[moto_4757_norm['batch'].isin(moto_4757_norm_stop.batch.unique())]
                                   .groupby('batch')['datetime'].max().reset_index()).copy()
# 获取开机稳定后时间
moto_data_4757_norm_stop_time['datetime_deta'] = moto_data_4757_norm_stop_time['datetime'] + timedelta(hours=1)
# 获取开机稳定后的batch
batch_deta = []
for i in moto_data_4757_norm_stop_time.datetime_deta.unique():
    batch_number = moto_4757_norm[moto_4757_norm['datetime'] == i].batch.values[0]
    batch_deta.append(batch_number)
moto_data_4757_norm_stop_time['batch_deta'] = batch_deta
moto_4757_norm['restart_flag'] = False
for i, j in zip(moto_data_4757_norm_stop_time.batch.values, moto_data_4757_norm_stop_time.batch_deta.values):
    moto_4757_norm['restart_flag'] = moto_4757_norm['restart_flag'] | ((moto_4757_norm.batch > i) & (moto_4757_norm.batch <= j))
# 筛选开机后的数据
moto_4757_norm_filter = moto_4757_norm[moto_4757_norm['restart_flag'] == False]
# 合并数据
moto_4757_whole_filter = pd.concat([moto_4757_norm_filter[['datetime', 'motor_id', 'motor_run', 'motor_speed', 'motor_temperature', 'motor_torque', 'sensor_state', 'label']], moto_4757_abnorm.dropna()], axis=0)
```

# 筛选motor_run,根据batch进行聚合

```python
moto_4757_whole = moto_4757_whole_filter
```

```python
# 获取batch标签
moto_4757_whole['batch'] = get_batch(moto_4757_whole, 'motor_run')
moto_4757_whole_run = moto_4757_whole[moto_4757_whole['motor_run'] == 1]
# 聚合方式
aggregations = {
    'motor_speed': ['min', 'max', 'mean', 'std', 'median'], 
    'motor_temperature': ['std', 'mean'], 
    'motor_torque': ['min', 'max', 'mean', 'std', 'median'],
    'sensor_state': ['mean', 'sum', 'std'],
    'datetime':['count']
}
moto_4757_whole_run_group = moto_4757_whole_run.groupby(['batch', 'label']).agg(aggregations).reset_index()
moto_4757_whole_run_group.columns = [i+'_'+j for i, j in moto_4757_whole_run_group.columns]
moto_4757_whole_run_group = moto_4757_whole_run_group.dropna()
```

# 对分组的数据保留异常数据以及正常数据运行时长大于20秒

```python
# 获取混淆矩阵
def validation(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    return cm
```

```python
def calculate_precision_recall_f1(cm):
    """
    计算精确率、召回率和F1值。

    参数:
    tp -- 真正例 (True Positive)
    fp -- 假正例 (False Positive)
    fn -- 假反例 (False Negative)

    返回:
    precision -- 精确率
    recall -- 召回率
    f1 -- F1值
    """
    tp = cm[1,1]
    fp = cm[1,0]
    fn = cm[0,1]
    
    # 计算精确率
    
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0

    # 计算召回率
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    # 计算F1值
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1
```

# 构建训练数据

```python
# 使用筛选后的数据进行建模
moto_4757_whole_run_group = filtered_df
# 训练数据
X_train = moto_4757_whole_run_group[moto_4757_whole_run_group['label_'] == 0]\
    [:int(0.95*len(moto_4757_whole_run_group))].drop(['label_', 'batch_'], axis=1)
# 验证数据
X_val_label = pd.concat([
    moto_4757_whole_run_group[moto_4757_whole_run_group['label_'] == 1],
    moto_4757_whole_run_group[moto_4757_whole_run_group['label_'] == 0][int(0.95*len(moto_4757_whole_run_group)):]
], axis=0)
# 验证数据标签
y_val = X_val_label['label_']
# 验证数据
X_val = X_val_label.drop(['label_', 'batch_'], axis=1)
# 特征缩放
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_val = scaler.fit_transform(X_val)
# 转换为PyTorch张量
X_train_tensor = torch.tensor(scaled_X_train).float()
X_val_tensor = torch.tensor(scaled_X_val).float()
```

# OneClassSVM

```python
from sklearn.svm import OneClassSVM
# 模型建立
ocsvm = OneClassSVM(kernel='rbf', gamma='auto', nu=0.03).fit(scaled_X_train)
# 模型验证
predictions = ocsvm.predict(scaled_X_val)
predictions = (predictions-1)/(-2)
cm = validation(y_val,predictions)
calculate_precision_recall_f1(cm)
```

# Iforest

```python
# 构建孤立森林
isolation_forest = IsolationForest(n_estimators=100, contamination='auto', random_state=42)
# 拟合模型并预测异常值分数
isolation_forest.fit(scaled_X_train)
predictions = isolation_forest.predict(scaled_X_val)
# predictions统一异常为1，正常为0
predictions = (predictions - 1)/-2
cm = validation(y_val,predictions)
calculate_precision_recall_f1(cm)
```
